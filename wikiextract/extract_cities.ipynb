Wiki Extract -- get city/gps data out of wikipedia

```python
import lxml
import pandas as pd    
page = 'https://en.wikipedia.org/wiki/List_of_cities_by_longitude'    
wikitables = pd.read_html(page, index_col=0, attrs={"class":"wikitable"})  
wikitables[0].to_csv('cities.csv')
```
