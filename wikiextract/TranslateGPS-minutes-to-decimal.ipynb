Translate GPS cordinates in the form of degrees and minutes into decimal degrees.

16°30′S,180°00′W
to -16.5, -180.0



```python
def degsecconv(degsec):
    direction = {'N':1, 'S':-1, 'E': 1, 'W':-1}
    n = degsec.replace(u'°',' ').replace(u'′',' ')
    n = n.split()
    d = n.pop()
    deg = int(n[0])
    fracdeg = int(n[1])/60.0
    strfracdeg = format(fracdeg,'.2f')
    signdeg = direction[d]
    return((deg+float(strfracdeg))*signdeg)
```

```python
import pandas as pd
df = pd.read_csv('cities.csv', sep=',', names=['Latitude', 'Longitude','City', 'Province/State', 'Country'],skiprows=2)
```

```python
d = dict()
for row in df.iterrows():
    num, e = row
    d[num] = [e['City'],degsecconv(e['Latitude']),degsecconv(e['Longitude'])]
    
df = pd.DataFrame.from_dict(d, orient="index")
df.to_csv('cities_degs.csv', index=False, header=['city','lat', 'lon'])

    
```
